Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Answer: Descriptive statistics summarize and describe data using measures such as mean,
median, mode, charts, and percentages. Inferential statistics use sample data to make predictions
or conclusions about a larger population. Example: Descriptive statistics can summarize the
average marks of a class. Inferential statistics can predict the performance of all students in a
school based on a sample.



Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Answer: Sampling is the process of selecting a subset of individuals from a population to study.
Random sampling gives every member an equal chance of being selected. Stratified sampling
divides the population into groups (strata) and samples from each group to ensure representation.


Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer: Mean is the average of values, median is the middle value after sorting, and mode is the most frequent value. These measures help summarize data and understand its central tendency.


Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Answer: Skewness measures the asymmetry of data distribution, while kurtosis measures the
peakedness or heaviness of tails. A positive skew means the distribution has a longer tail on the right side, indicating some high values pull the distribution upward.



Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
import statistics
import numpy as np
mean_value = np.mean(numbers)
median_value = np.median(numbers)
mode_value = statistics.multimode(numbers)
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
# Output:
# Mean: 19.6
# Median: 19.0
# Mode: [12, 19, 24]
```

Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]


```python
    import numpy as np
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
covariance = np.cov(list_x, list_y)[0,1]
correlation = np.corrcoef(list_x, list_y)[0,1]
print("Covariance:", covariance)
print("Correlation:", correlation)
# Output:
# Covariance: 275.0
# Correlation: 0.995893206467704
```

Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


```python
# Outliers identified using IQR method: [35]
import matplotlib.pyplot as plt
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
plt.boxplot(data)
plt.title("Boxplot")
plt.show()
```

Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
● Explain how you would use covariance and correlation to explore this relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

 Covariance helps determine whether advertising spend and sales move together, while correlation measures the strength and direction of the relationship. A positive correlation suggests that higher advertising spend is associated with higher sales. Computed correlation = 0.9936.

```python
import numpy as np
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]
correlation = np.corrcoef(advertising_spend, daily_sales)[0,1]
print("Correlation:", correlation)
# Output:
# Correlation: 0.9935824101653329
```

Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


Summary statistics such as mean, median, standard deviation, minimum, and maximum should be used to understand the survey data. A histogram helps visualize the distribution, spread, and concentration of scores.


```python
import matplotlib.pyplot as plt
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
plt.hist(survey_scores)
plt.title("Histogram of Survey Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
